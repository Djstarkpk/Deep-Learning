```python
import pandas as pd
import re
from googleapiclient.discovery import build
from datetime import datetime

# ============================================
# CONFIGURATION
# ============================================
# Replace with your YouTube Data API key
API_KEY = 'AIzaSyAxuWWNJNAmeOREpIuudkq9virEgYZLtFk'

# ============================================
# HELPER FUNCTIONS
# ============================================

def extract_video_id(url):
    """
    Extract video ID from various YouTube URL formats
    """
    if not url or pd.isna(url):
        return None

    # Handle different YouTube URL formats
    patterns = [
        r'(?:youtube\.com\/watch\?v=|youtu\.be\/|youtube\.com\/embed\/)([^&\n?#]+)',
        r'youtube\.com\/watch\?.*v=([^&\n?#]+)'
    ]

    for pattern in patterns:
        match = re.search(pattern, str(url))
        if match:
            return match.group(1)

    return None

def get_video_details(video_id, youtube):
    """
    Fetch video details using YouTube Data API
    """
    if not video_id:
        return None, None

    try:
        request = youtube.videos().list(
            part='snippet',
            id=video_id
        )
        response = request.execute()

        if response['items']:
            snippet = response['items'][0]['snippet']

            # Extract publish date
            publish_date = snippet['publishedAt']
            # Convert to readable format
            date_obj = datetime.strptime(publish_date, '%Y-%m-%dT%H:%M:%SZ')
            formatted_date = date_obj.strftime('%Y-%m-%d')

            # Extract description
            description = snippet['description']

            return formatted_date, description
        else:
            return None, None

    except Exception as e:
        print(f"Error fetching video {video_id}: {str(e)}")
        return None, None

# ============================================
# MAIN FUNCTION
# ============================================

def add_youtube_data(df, url_column='Youtube URL'):
    """
    Add release date and description columns to the dataframe

    Parameters:
    -----------
    df : pandas.DataFrame
        Input dataframe with YouTube URLs
    url_column : str
        Name of the column containing YouTube URLs

    Returns:
    --------
    pandas.DataFrame
        Updated dataframe with new columns
    """

    # Initialize YouTube API client
    youtube = build('youtube', 'v3', developerKey=API_KEY)

    # Initialize new columns
    release_dates = []
    descriptions = []

    # Process each URL
    for idx, row in df.iterrows():
        url = row[url_column]
        print(f"Processing row {idx + 1}/{len(df)}: {url}")

        # Extract video ID
        video_id = extract_video_id(url)

        # Get video details
        release_date, description = get_video_details(video_id, youtube)

        release_dates.append(release_date)
        descriptions.append(description)

    # Add new columns to dataframe
    df['Release Date'] = release_dates
    df['Video Description'] = descriptions

    return df

# ============================================
# EXAMPLE USAGE
# ============================================

if __name__ == "__main__":

    # Add YouTube data
    df_updated = add_youtube_data(df)

    print("\nUpdated DataFrame:")
    print(df_updated)

    # Save to CSV if needed
    # df_updated.to_csv('updated_data.csv', index=False)
```

```python
def clean_title(title):
  keywords=['chief','cmo','cto','ceo','svp','evp','president']
  title=title.lower()
  return any(k in title for k in keywords)
def description_matches(row):
  desc=str(row['Video Description']).lower()
  name=str(row['Name']).lower()
  title=str(row['Title']).lower()
  title_keywords=['chief','cmo','cto','ceo','svp','evp','president']
  title_match=any(k in desc and k in title for k in title_keywords)
  name_match= name in desc
  return name_match or title_match
df=df[df.apply(description_matches,axis=1)]
```

```python
def title_priority(title):
  title=title.lower()
  if 'ceo' in title or 'president' in title:
    return 1
  elif 'chief marketing' in title or 'cmo' in title or 'cro' in title or 'commercial' in title:
    return 2
  elif 'cto' in title or 'cio' in title or 'cpo'  in title:
    return 3
  elif 'evp' in title or 'executive vice president' in title:
    return 4
  elif 'svp' in title or 'senior vice president' in title:
    return 5
  else:
    return 99

df_latest_exec['title_priority']=df_latest_exec['Title'].apply(title_priority)
df_latest_exec=df_latest_exec.sort_values(by=['title_priority','Release Date'],ascending=[True ,False])
df_final_50=df_latest_exec.head(50).reset_index(drop=True)

```
